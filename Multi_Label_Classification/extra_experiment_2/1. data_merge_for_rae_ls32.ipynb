```python
import pandas as pd
```

```python
version = 'v2'
```

```python
train_abst = pd.read_excel('./data/paper_train.xlsx')
train_abst = pd.DataFrame(train_abst['abstract'])

val_abst = pd.read_excel('./data/paper_val.xlsx')
val_abst = pd.DataFrame(val_abst['abstract'])

test_abst = pd.read_excel('./data/paper_test.xlsx')
test_abst = pd.DataFrame(test_abst['abstract'])



train_label = pd.read_excel('./data/rae_ls32_{}_latent_vector_train.xlsx'.format(version))
train_label = train_label.drop(['Unnamed: 0'], axis=1)

val_label = pd.read_excel('./data/rae_ls32_{}_latent_vector_val.xlsx'.format(version))
val_label = val_label.drop(['Unnamed: 0'], axis=1)

test_label = pd.read_excel('./data/rae_ls32_{}_latent_vector_test.xlsx'.format(version))
test_label = test_label.drop(['Unnamed: 0'], axis=1)
```
