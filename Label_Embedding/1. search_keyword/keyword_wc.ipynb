```python
#   (TSV) wordcloud  

import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from PIL import Image


FONTPATH =  "./font/NanumGothic.ttf"                       #  
input_file_name = "./keyword_count.txt"   #   
IMG_MASK_PATH = "./engin5_mask.png"             #   
SAVE_IMG_PATH = "./results_wc.png"                     #    
#USER_STOPWORDS = ["", ""]                      #  
MAX_WORDS = 500                                                #    
MAX_FONT = 150                                                 #     


def read_input(input_file_name):
     
    data = {}
    
    with open(input_file_name, "r", encoding="utf-8") as input_file:
        for line_num, line in enumerate(input_file):
            line = line.strip()
            cols = line.split("\t")
            if cols[0] not in USER_STOPWORDS:
                data[cols[0]] = int(cols[1])
   
    return data


data = read_input(input_file_name)
img_mask = np.array(Image.open(IMG_MASK_PATH))   
image_colors = ImageColorGenerator(img_mask)

                                                                
wordcloud = WordCloud(max_font_size = MAX_FONT, max_words=MAX_WORDS, random_state = 24, 
               mask = img_mask, background_color = "white", font_path = FONTPATH)


wc = wordcloud.generate_from_frequencies(data)

plt.figure(figsize = (15, 15))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.savefig(SAVE_IMG_PATH )                      
plt.show()
```
