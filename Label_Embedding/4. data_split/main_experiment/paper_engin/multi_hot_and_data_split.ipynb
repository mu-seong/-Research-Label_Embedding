# multi hot encoding

```python
data['keywords'] = pd.Series()

for idx, key in tqdm(enumerate(data.keyword_count)):
    tmp = key.replace('[', '').replace(']', '').split(', ')
    tmp_3 = []
    for tmp_2 in tmp:
        tmp_3.append(tmp_2.replace("'", ""))
        tmp_3 = list(filter(None, tmp_3))
    #print(tmp_3)
    data['keywords'].loc[idx] = tmp_3
```

```python
data
```

```python
data['keywords'][11]
```

```python
multilabel_binarizer = MultiLabelBinarizer()
multilabel_binarizer.fit(data.keywords)
y = multilabel_binarizer.classes_
```

```python
y
```

```python
len(y)
```

```python
def to_multi_hot(data, y):
    vocab = {}
    
    for idx, label_name in tqdm(enumerate(y)):
        vocab[label_name] = idx
    
    label_array = np.zeros((len(data), len(y)))
    
    for index in tqdm(range(len(data))):
        for label in y:
            label_idx = vocab.get(label)
        
            if label in data.keywords[index]:
                label_array[index, label_idx] = 1
        
    return label_array
```

```python
multi_hot_labels = to_multi_hot(data, y)
multi_hot_labels
```

```python
paper_labels = pd.DataFrame(multi_hot_labels)
paper_labels
```
